### TASK-4 : Image to Pencil Sketch with Python:

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python

# Importing The Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

# Read the Image

```python
original_img = cv2.imread('dog.jpg')
```

```python
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(original_img))
```

# Converting the Original Image into Grayscale

```python
gray_image = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# Inverting the Grayscale image

```python
inv_gray_image = 255 - gray_image
```

```python
inv_gray_image
```

```python
display(Image.fromarray(inv_gray_image))
```

# Blur the image

```python
blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
```

```python
display(Image.fromarray(blur_image))
```

# Mixing the grayscale image and inverted blur image

```python
inv_blur_img = 255 - blur_image
    
```

```python
def dodge(gray_image, inv_blur_img):
    return cv2.divide(gray_image, inv_blur_img, scale=250.0)
```

```python
pencil_sketch = dodge(gray_image, inv_blur_img)
```

```python
display(Image.fromarray(pencil_sketch))
```
